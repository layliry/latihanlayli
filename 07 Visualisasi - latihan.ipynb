Visualisasi Data

```python
# import package python yang akan digunakan

import numpy as py
import pandas as pd

%matplotlib inline
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
# import data

df_fin = pd.read_excel("Finland.xlsx", 
                        # ambil sheet yang akan digunakan
                        sheet_name = "Finland by Citizenship",
                        # hapus header yang ga dibutuhkan
                        skiprows=20,
                        # hapus baris terakhir
                        skipfooter=2,
                        #cleaning unknown value
                        na_values=".."
                        )
df_fin
```

```python
# Eksplorasi Data

df_fin.info()
```

```python
df_fin.describe()
```

# 1. Line Plot

```python
# Trend imigran indonesia dari 1980 s.d. 2013
df_line = df_fin[df_fin["OdName"]=="Indonesia"]
df_line
```

```python
df_line = df_line.tail(1)
df_line
```

```python
# coba plot
df_line.plot()

# hasil terbalik x menjadi y, y menjadi x.
```

```python
# Transpose
df_line.T
```

```python
# ambil data pada tahun

tahun = range(1980,2014)
```

```python
# transpose dan plot
df_line = df_line[tahun].T

```

```python
df_line.plot(figsize=(13,5),marker="o")

plt.title('OK')
```
